## Simple XOR

This notebook implements a Genetic Algorithm (GA) that evolves a circuit based on SN P-System gates using a the XOR funciton as a reference.

This example is considered simple because it uses the high-level API offered by the library, making it easy (only few lines) to find such a circuit.


### 1) Loading the required libraries

The external required libraries are:
- Plots
- DataFrames
- Random

The last two are loaded by SpikingNeuralEvolution


```julia
# This line allows to load the library even if it is not installed in the system
push!(LOAD_PATH, "../src/")

using SpikingNeuralEvolution
using Plots
```

## 2) Choosing the input space

We simply define a value $n$ which will control the magnitude of the input space. Do not use large $n$, since the required space is exponential with respect to $n$.

```julia
n = UInt16(4);
```

This means that we will use $2^4 = 16$ possible combinations of bits to evaluate the circuit.

## 3) Defining the Boolean function

Now we define the target Boolean function, that will simply be the XOR of the input bits

```julia
function XOR(inputs::Vector{Bool}) 
    return reduce(xor, inputs)
end
;
```

## 4) Evolve the population

With all these simple ingredients, we are able to launch a simulation.

```julia
histories = Evolve(XOR, n);
```

## 5) Visualize the results

It is possible to plot the history of fitness values

```julia
plot(histories)
```
